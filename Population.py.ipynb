```python
import pandas as pd
import numpy as np
df = pd.read_excel('population_mondiale.xls') 
df.head(10) 
df.tail(5)
df.columns
df['Country Name'].count()
df.info()
df.isnull().sum()
df[df.isnull().any(axis=1)]
df = df.drop(columns=['Indicator Name', 'Indicator Code'])
df.head()
df = df.dropna()
df.isnull().sum() 
df.info()
df[df['Country Name'] == 'Tunisie']  
df1 = df.sort_values(by='Country Name', ascending=True)
print(df1.head(20))  
top_10_2024 = df[['Country Name', 2024]].sort_values(by=2024, ascending=False).head(10)
print(top_10_2024) 
bottom_10_2024 = df[['Country Name', 2024]].sort_values(by=2024, ascending=True).head(10)
print(bottom_10_2024)

import matplotlib.pyplot as plt
df[df['Country Name'] == 'Tunisie']  
tunisia_data = df[df['Country Name'] == 'Tunisie'].iloc[0, 2:]  
years = tunisia_data.index.astype(int)
plt.plot(years, tunisia_data.values, label='Tunise', color='red', linestyle='--')

plt.title('Évolution de la population de la Tunisie (1960-2024)')
plt.xlabel('Année')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()


df_china_india = df[df['Country Name'].isin(['Chine', 'Inde'])] 

d1=df[df['Country Name'] == 'Chine'].iloc[0, 2:]
y = d1.index.astype(int)
d2=df[df['Country Name'] == 'Inde'].iloc[0, 2:]

plt.scatter(y, d1.values, label='Chine', color='red', linestyle='--')
plt.scatter(y, d2.values, label='Inde', color='blue', linestyle='-.')
plt.title('Évolution de la population de la Chine et de l\'Inde (1960-2024)')
plt.xlabel('Année')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()
```
